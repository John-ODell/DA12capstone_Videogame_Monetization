```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import html5lib
```

```python
URL = 'https://vgsales.fandom.com/wiki/List_of_best-selling_video_games'

response = requests.get(URL)
```

```python
html_content = response.content
soup = BS(html_content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find('title').text
```

```python
first_paragraph = soup.find('p')
```

```python
print(first_paragraph)
```

```python
table = soup.find('table', {'class': 'wikitable'})
```

```python
best_selling_vedeo_games = str(soup.find('table', attrs={'class' : 'wikitable'}))

from IPython.core.display import HTML

HTML(best_selling_vedeo_games)
```

```python
best_selling_vedeo_games = pd.read_html(str(soup.find('table', attrs={'class' : 'wikitable'})))[0]
```

```python
best_selling_vedeo_games
```

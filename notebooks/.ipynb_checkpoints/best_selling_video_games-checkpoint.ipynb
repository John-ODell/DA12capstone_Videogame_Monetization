```python
import requests
from bs4 import BeautifulSoup as BS
import pandas as pd
import html5lib
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
file_path = r'C:\Users\johnc\Desktop\DA12\Python\DA12capstone_Videogame_Monetization\data\best_selling_video_games_raw.csv'
best_selling_video_games_raw = pd.read_csv(file_path)
```

```python
best_selling_video_games = best_selling_video_games_raw
```

```python
best_selling_video_games = best_selling_video_games.drop(columns = 'Ref')
```

```python
best_selling_video_games['General_Platform']=best_selling_video_games['Platform']
```

```python
best_selling_video_games['Company_Platform']=best_selling_video_games['Platform']
```

```python
best_selling_video_games.General_Platform.value_counts()
```

```python
mobile_dictionary = {'Android':'Mobile','iOS':'Mobile'}
```

```python
handheld_dictionary = {'DS':'Handheld','Game Boy':'Handheld','GBC':'Handheld','GBA':'Handheld','3DS':'Handheld'}
```

```python
console_dictionary = {'Switch':'Console','Wii':'Console','NES':'Console','X360':'Console','PS3':'Console','PS4':'Console','Mega Drive':'Console','SNES':'Console'}
```

```python
released_on_dictionary = {**mobile_dictionary, **handheld_dictionary, **console_dictionary}
```

```python
nintendo_dictionary =  {'Switch':'Nintendo','Wii':'Nintendo','DS':'Nintendo','NES':'Nintendo','Game Boy':'Nintendo','GBC':'Nintendo','GBA':'Nintendo','SNES':'Nintendo','3DS':'Nintendo','Mega Drive':'Nintendo'}
```

```python
sony_dictionary = {'PS3':'Sony','PS4':'Sony'}
```

```python
microsoft_dictionary = {'XBOX':'Microsoft','X360':'Microsoft'}
```

```python
apple_dictionary = {'iOS':'Apple'}
```

```python
companies_dictionary = {**nintendo_dictionary,**sony_dictionary,**microsoft_dictionary,**apple_dictionary}
```

```python
best_selling_video_games['General_Platform']=best_selling_video_games['General_Platform'].replace(released_on_dictionary)
```

```python
best_selling_video_games['Company_Platform']=best_selling_video_games['Company_Platform'].replace(companies_dictionary)
```

## droped the 2005 packman, the console did not make sense.

```python
best_selling_video_games = best_selling_video_games.drop(index=14)
```

```python
plt.figure(figsize=(10,6))
ax=sns.scatterplot(data=best_selling_video_games,
                x='Year',
                y='General_Platform',
                hue='Company_Platform',
                palette=['red','green','blue','black','brown','yellow'])

plt.xticks(rotation=45, ha='right')

for line in range(0, best_selling_video_games.shape[0]):
    # Alternate the position of the annotations
    if line % 2 == 0:
        xytext = (-5, 15)  # Above the plot
    else:
        xytext = (5, -15)  # Below the plot

    ax.annotate(best_selling_video_games.Title.iloc[line],
                (best_selling_video_games.Year.iloc[line], best_selling_video_games.General_Platform.iloc[line]),
                textcoords="offset points",
                xytext=xytext,
                ha='left',  
                rotation=45,  
                fontsize='medium',
                color='black',
                weight='light')


# Show the plot
#plt.savefig(r'C:\Users\johnc\Desktop\DA12\Python\DA12capstone_Videogame_Monetization\data\Console_Model_unitssold_type.png', bbox_inches='tight')
plt.show()
```

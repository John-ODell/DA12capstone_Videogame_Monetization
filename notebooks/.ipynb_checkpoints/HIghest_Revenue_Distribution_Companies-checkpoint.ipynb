```python
def convert_to_int(value):
    value = value.replace('$', '').replace(',', '')
    if 'billion' in value:
        return int(float(value.replace(' billion', '')) * 1_000_000_000)
    elif 'million' in value:
        return int(float(value.replace(' million', '')) * 1_000_000)
    else:
        return int(value)
```

```python
highest_revenue_companies['Revenue_Int'] = highest_revenue_companies['Revenue'].apply(convert_to_int)
```

```python
def clean_units_sold(value):
    value = value.replace('\xa0', ' ')
    value = value.replace('>', '').replace('~', '').replace(' million', '').replace('(estimate)', '').strip()
    
    if '–' in value:
        range_values = value.split('–')
        value = (float(range_values[0]) + float(range_values[1])) / 2
    
    return int(float(value) * 1e6)
```
